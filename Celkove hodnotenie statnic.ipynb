# Program na celkové hodnotenie štátnic

```python
import tkinter as tk  # Importuje modul tkinter a priradí mu skratku tk
from tkinter import filedialog  # Importuje funkciu filedialog zo modulu tkinter
import pandas as pd  # Importuje modul pandas a priradí mu skratku pd

def vypocitat_priemer():
    # Získanie vstupu z textového poľa pre meno študenta
    meno = entry_meno.get()
    if not meno:
        label_vysledok.config(text="Chyba: Zadajte meno študenta.")
        return

    # Získanie vstupov zo všetkých textových polí pre známky a konverzia na veľké písmená
    BP = entry_bp.get().upper()
    GEO = entry_geo.get().upper()
    GNS = entry_gns.get().upper()
    PA = entry_pa.get().upper()

    while BP not in hodnoty and BP != '':
        # Kontrola platnosti známky pre Bakalárska práca a jej obhajoba
        entry_bp.delete(0, tk.END)
        entry_bp.insert(tk.END, "Neplatná znamka. Zadajte znova.")
        return

    while GEO not in hodnoty and GEO != '':
        # Kontrola platnosti známky pre Geodézia
        entry_geo.delete(0, tk.END)
        entry_geo.insert(tk.END, "Neplatná znamka. Zadajte znova.")
        return

    while GNS not in hodnoty and GNS != '':
        # Kontrola platnosti známky pre Globálne geodetické systémy
        entry_gns.delete(0, tk.END)
        entry_gns.insert(tk.END, "Neplatná znamka. Zadajte znova.")
        return

    while PA not in hodnoty and PA != '':
        # Kontrola platnosti známky pre Priestorová analýza
        entry_pa.delete(0, tk.END)
        entry_pa.insert(tk.END, "Neplatná znamka. Zadajte znova.")
        return

    # Konverzia zadaných známok na ich hodnoty
    BP = hodnoty.get(BP, None)
    GEO = hodnoty.get(GEO, None)
    GNS = hodnoty.get(GNS, None)
    PA = hodnoty.get(PA, None)

    # Vytvorenie zoznamu hodnôt známok, pričom ignoruje hodnoty None
    hodnoty_list = [BP, GEO, GNS, PA]
    hodnoty_list = [hodnota for hodnota in hodnoty_list if hodnota is not None]

    # Výpočet priemeru hodnôt známok
    priemer = sum(hodnoty_list) / len(hodnoty_list) if hodnoty_list else 0

    if "FX" in hodnoty_list:
        vysledok = "Nedostatočne"
    elif priemer <= 1.5 and hodnoty_list.count("C") <= 1 and not any(hodnota in ["D", "E"] for hodnota in hodnoty_list):
        vysledok = "Výborne"
    else:
        vysledok = "Vyhovel(a)"
    
    # Zobrazenie výsledkov na obrazovke
    label_priemer.config(text="Priemer známok: {:.2f}".format(priemer))
    label_hodnotenie.config(text="Hodnotenie: " + vysledok)

def export_do_excelu():
    # Získanie vstupov zo všetkých textových polí a zobrazenie výsledkov z labelov
    meno = entry_meno.get()
    vstup_bp = entry_bp.get()
    vstup_geo = entry_geo.get()
    vstup_gns = entry_gns.get()
    vstup_pa = entry_pa.get()
    priemer = label_priemer["text"].split(": ")[1]
    hodnotenie = label_hodnotenie["text"].split(": ")[1]
    
    # Vytvorenie slovníka s údajmi pre export do Excelu
    data = {
        "Meno a priezvisko": [meno],
        "Bakalárska práca a jej obhajoba": [vstup_bp],
        "Geodézia": [vstup_geo],
        "Globálne geodetické systémy": [vstup_gns],
        "Priestorová analýza": [vstup_pa],
        "Priemer": [priemer],
        "Hodnotenie": [hodnotenie]
    }

    # Vytvorenie DataFrame z údajov
    df = pd.DataFrame(data)
    
    # Zobrazenie dialógového okna pre uloženie súboru Excelu
    file_path = filedialog.asksaveasfilename(defaultextension=".xlsx")
    if file_path:
        # Exportovanie DataFrame do Excelu
        df.to_excel(file_path, index=False)
        label_export.config(text="Export do Excelu dokončený.")

root = tk.Tk()  # Vytvorenie okna aplikácie pomocou triedy Tk z modulu tkinter
root.title("Hodnotenie štátnej skúšky z odboru Geodézia a kartografia")  # Nastavenie titulku okna

hodnoty = {
    'A': 1,
    'B': 1.5,
    'C': 2,
    'D': 2.5,
    'E': 3,
    'FX': 4
}

# Vytvorenie a umiestnenie grafických komponentov v okne aplikácie
label_meno = tk.Label(root, text="Meno a priezvisko študenta:")
label_meno.pack()
entry_meno = tk.Entry(root)
entry_meno.pack()

label_bp = tk.Label(root, text="Bakalárska práca a jej obhajoba:")
label_bp.pack()
entry_bp = tk.Entry(root)
entry_bp.pack()

label_geo = tk.Label(root, text="Geodézia:")
label_geo.pack()
entry_geo = tk.Entry(root)
entry_geo.pack()

label_gns = tk.Label(root, text="Globálne geodetické systémy:")
label_gns.pack()
entry_gns = tk.Entry(root)
entry_gns.pack()

label_pa = tk.Label(root, text="Priestorová analýza:")
label_pa.pack()
entry_pa = tk.Entry(root)
entry_pa.pack()

button_vypocitat = tk.Button(root, text="Vypočítať priemer", command=vypocitat_priemer)
button_vypocitat.pack()

button_export = tk.Button(root, text="Exportovať do Excelu", command=export_do_excelu)
button_export.pack()

label_vysledok = tk.Label(root, text="")
label_vysledok.pack()

label_priemer = tk.Label(root, text="")
label_priemer.pack()

label_hodnotenie = tk.Label(root, text="")
label_hodnotenie.pack()

label_export = tk.Label(root, text="")
label_export.pack()

root.geometry("490x450")  # Nastavenie veľkosti okna

root.mainloop()  # Spustenie hlavnej smyčky pre spracovanie udalostí v aplikácii
```

## Podrobný popis riadkov programu

```python

import tkinter as tk
#Tento riadok importuje modul `tkinter` a priradzuje mu skratku `tk`. Modul `tkinter` poskytuje nástroje na vytváranie grafického užívateľského rozhrania.

from tkinter import filedialog
#Tento riadok importuje konkrétnu funkciu `filedialog` z modulu `tkinter`. Táto funkcia poskytuje dialógové okná pre prácu so súbormi (napríklad otváranie, ukladanie súborov).

import pandas as pd
#Tento riadok importuje modul `pandas` a priradzuje mu skratku `pd`. Modul `pandas` poskytuje vysokoúrovňové dátové štruktúry a nástroje na analýzu a manipuláciu s dátami.

def vypocitat_priemer():
#Týmto riadkom začína definícia funkcie `vypocitat_priemer()`. Táto funkcia bude vypočítavať priemer hodnôt známok na základe vstupných hodnôt zo vstupných polí.

    meno = entry_meno.get()
    #Tento riadok získava vstupnú hodnotu z textového poľa `entry_meno` (ktoré obsahuje meno študenta) a ukladá ju do premennej `meno`.

    if not meno:
        label_vysledok.config(text="Chyba: Zadajte meno študenta.")
    return
#Tento blok kódu overuje, či je premenná `meno` prázdna. Ak je prázdna, znamená to, že študent nezadal meno, a v tom prípade sa na štandardný výstupový label `label_vysledok` zobrazí chybová správa "Chyba: Zadajte meno študenta." a funkcia sa ukončí pomocou `return`, čím sa zastaví ďalšie vykonávanie funkcie.

BP = entry_bp.get().upper()
GEO = entry_geo.get().upper()
GNS = entry_gns.get().upper()
PA = entry_pa.get().upper()
#Tieto riadky získavajú vstupné hodnoty z textových polí `entry_bp`, `entry_geo`, `entry_gns` a `entry_pa` (ktoré obsahujú známky) a ukladajú ich do premenných `BP`, `GEO`, `GNS` a `PA` po prevedení na veľké písmená pomocou metódy `upper()`. Tým sa zabezpečí jednotný formát zadaných známok.

while BP not in hodnoty and BP != '':
    entry_bp.delete(0, tk.END)
    entry_bp.insert(tk.END, "Neplatná znamka. Zadajte znova.")
    return
#Tento blok kódu overuje, či zadaná hodnota `BP` je platnou známkou. Pokiaľ nie je, odstráni sa obsah z textového poľa `entry_bp`, vloží sa chybová správa "Neplatná znamka. Zadajte znova." a funkcia sa ukončí.
#Podobný blok kódu sa opakuje pre premenné `GEO`, `GNS` a `PA` s príslušnými textovými poliami.

BP = hodnoty.get(BP, None)
GEO = hodnoty.get(GEO, None)
GNS = hodnoty.get(GNS, None)
PA = hodnoty.get(PA, None)
#Tieto riadky mapujú hodnoty známok (`BP`, `GEO`, `GNS`, `PA`) na ich ekvivalentné numerické hodnoty pomocou slovníka `hodnoty`. Ak zadaná hodnota nie je v slovníku, použije sa hodnota `None`.

hodnoty_list = [BP, GEO, GNS, PA]
hodnoty_list = [hodnota for hodnota in hodnoty_list if hodnota is not None]
#Tieto riadky vytvárajú zoznam `hodnoty_list`, ktorý obsahuje numerické hodnoty známok (`BP`, `GEO`, `GNS`, `PA`), pričom sa odstráni všetky hodnoty `None` z tohto zoznamu.

priemer = sum(hodnoty_list) / len(hodnoty_list) if hodnoty_list else 0
#Tento riadok vypočíta priemer hodnôt v zozname `hodnoty_list`. Ak je zoznam prázdny (neobsahuje žiadne hodnoty), priemer sa nastaví na hodnotu 0.

if "FX" in hodnoty_list:
    vysledok = "Nedostatočne"
elif priemer <= 1.5 and hodnoty_list.count("C") <= 1 and not any(hodnota in ["D", "E"] for hodnota in hodnoty_list):
    vysledok = "Výborne"
else:
    vysledok = "Vyhovel(a)"
#Tento blok kódu určuje výsledok hodnotenia na základe priemeru a hodnôt známok. Ak sa v zozname `hodnoty_list` nachádza hodnota "FX", výsledkom je "Nedostatočne". Ak je priemer menší alebo rovný 1.5, obsahuje najviac jednu hodnotu "C" a neobsahuje žiadnu hodnotu "D" ani "E", výsledkom je "Výborne". V opačnom prípade je výsledkom "Vyhovel(a)".

label_priemer.config(text="Priemer známok: {:.2f}".format(priemer))
label_hodnotenie.config(text="Hodnotenie: " + vysledok)
#Tieto riadky nastavujú textové hodnoty pre labely `label_priemer` a `label_hodnotenie`. V `label_priemer` sa zobrazí hodnota priemeru známok s presnosťou na dve desatinné miesta. V `label_hodnotenie` sa zobrazí výsledok hodnotenia.

def export_do_excelu():
#Týmto riadkom začína definícia funkcie `export_do_excelu()`. Táto funkcia bude exportovať údaje do Excelu.

    meno = entry_meno.get()
    vstup_bp = entry_bp.get()
    vstup_geo = entry_geo.get()
    vstup_gns = entry_gns.get()
    vstup_pa = entry_pa.get()
    priemer = label_priemer["text"].split(": ")[1]
    hodnotenie = label_hodnotenie["text"].split(": ")[1]
    #Tieto riadky získavajú hodnoty z textových polí (`entry_meno`, `entry_bp`, `entry_geo`, `entry_gns`, `entry_pa`) a z labelov (`label_priemer`, `label_hodnotenie`) a ukladajú ich do príslušných premenných (`meno`, `vstup_bp`, `vstup_geo`, `vstup_gns`, `vstup_pa`, `priemer`, `hodnotenie`).

data = {
    "Meno a priezvisko": [meno],
    "Bakalárska práca a jej obhajoba": [vstup_bp],
    "Geodézia": [vstup_geo],
    "Globálne geodetické systémy": [vstup_gns],
    "Priestorová analýza": [vstup_pa],
    "Priemer": [priemer],
    "Hodnotenie": [hodnotenie]
}
#Tento riadok vytvára slovník `data`, kde kľúčami sú názvy stĺpcov a hodnotami sú zoznamy s príslušnými hodnotami (`meno`, `vstup_bp`, `vstup_geo`, `vstup_gns`, `vstup_pa`, `priemer`, `hodnotenie`).

df = pd.DataFrame(data)
#Tento riadok vytvára dátový rámec (`DataFrame`) `df` z údajov v slovníku `data` pomocou modulu `pandas`.


file_path = filedialog.asksaveasfilename(defaultextension=".xlsx")
if file_path:
    df.to_excel(file_path, index=False)
    label_export.config(text="Export do Excelu dokončený.")
#Tieto riadky otvárajú dialógové okno pre ukladanie súboru pomocou funkcie `asksaveasfilename` z modulu `filedialog`. Ak je vybraný súbor a bol zadaný názov, dátový rámec `df` sa exportuje do Excelu s použitím funkcie `to_excel`, pričom sa nastaví cesta k
#súboru `file_path` a vypíše sa správa "Export do Excelu dokončený" na label `label_export`.


window = tk.Tk()
#Tento riadok vytvára hlavné okno aplikácie pomocou konštruktora `Tk` z modulu `tkinter`.


window.title("Hodnotenie študenta")
#Tento riadok nastavuje titulok hlavného okna na "Hodnotenie študenta".

label_meno = tk.Label(window, text="Meno študenta:")
#Tento riadok vytvára label `label_meno` v hlavnom okne s textom "Meno študenta:".


entry_meno = tk.Entry(window)

#Tento riadok vytvára textové pole `entry_meno` v hlavnom okne, do ktorého môže študent zadať svoje meno.
#Podobne sa postupuje aj pre labely a textové polia pre hodnoty známok `BP`, `GEO`, `GNS` a `PA`.


button_priemer = tk.Button(window, text="Vypočítať priemer", command=vypocitat_priemer)
#Tento riadok vytvára tlačidlo `button_priemer` s textom "Vypočítať priemer" v hlavnom okne. Toto tlačidlo spúšťa funkciu `vypocitat_priemer` po kliknutí.


button_export = tk.Button(window, text="Exportovať do Excelu", command=export_do_excelu)
#Tento riadok vytvára tlačidlo `button_export` s textom "Exportovať do Excelu" v hlavnom okne. Toto tlačidlo spúšťa funkciu `export_do_excelu` po kliknutí.


label_priemer = tk.Label(window, text="Priemer známok:")
#Tento riadok vytvára label `label_priemer` v hlavnom okne s textom "Priemer známok:".


label_hodnotenie = tk.Label(window, text="Hodnotenie:")
#Tento riadok vytvára label `label_hodnotenie` v hlavnom okne s textom "Hodnotenie:".


label_export = tk.Label(window, text="")
#Tento riadok vytvára label `label_export` v hlavnom okne s prázdny textom.


label_meno.grid(row=0, column=0, sticky=tk.W)
entry_meno.grid(row=0, column=1)
#Tieto riadky umiestňujú label `label_meno` a textové pole `entry_meno` na prvý riadok a prvé dve stĺpce hlavného okna. `sticky=tk.W` znamená, že label a textové pole budú zarovnané vľavo.
#Podobne sa postupuje aj pre labely a textové polia pre hodnoty známok `BP`, `GEO`, `GNS` a  `PA`.


button_priemer.grid(row=5, column=0, columnspan=2)
button_export.grid(row=6, column=0, columnspan=2)
#Tieto riadky umiestňujú tlačidlá `button_priemer` a `button_export` na piaty a šiesty riadok a do oboch stĺpcov hlavného okna. `columnspan=2` znamená, že tlačidlá rozpínajú cez oba stĺpce.


label_priemer.grid(row=7, column=0, sticky=tk.W)
label_hodnotenie.grid(row=8, column=0, sticky=tk.W)
label_export.grid(row=9, column=0, columnspan=2)
#Tieto riadky umiestňujú labely `label_priemer`, `label_hodnotenie` a `label_export` na siedmy, ôsmy a deviaty riadok a do prvých dvoch stĺpcov hlavného okna.


window.mainloop()
#Tento riadok spúšťa hlavnú smyčku aplikácie, ktorá čaká na udalosti a reaguje na ne, ako sú kliknutia na tlačidlá alebo zmeny vstupov.
```
